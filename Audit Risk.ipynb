```python
from sklearn.model_selection import train_test_split,cross_val_score
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2,stratify=Y, random_state = 123)
```

# lets build our model we gonna try 3 different algorithms for this project 
# our base model  Logistic model it is the benchmark fo any classification problem its better we try it out first and go for other algorithms, 
# in Logistic Regression we use Lasso regression as penality factor,. lasso shrinks less important features and removes permenatly from dataset thus we have features best represent our dataset

```python
# lest import our model
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()
reg.fit(X_train_scaled, Y_train)
```

```python
# lets perform cross validation our dataset
```

```python
from sklearn import model_selection 
from sklearn.metrics import accuracy_score,  roc_auc_score
from sklearn.model_selection import KFold
kf = KFold(n_splits=10)
kf.get_n_splits(X)
cross_val = KFold(n_splits=10, random_state=None, shuffle=False)

scoring = 'accuracy'

acc_reg = cross_val_score(estimator = reg, X = X_train_scaled, y = Y_train, cv = kf, scoring=scoring)
acc_reg.mean()
```

 above we imported kfold cross validation it is amodel validation technique when delaing with small datasets
When a specific value for k is chosen, it may be used in place of k in the reference to the model, such as k=10 becoming 10-fold cross-validation.
Shuffle the dataset randomly.
Split the dataset into k groups
For each unique group:
Take the group as a hold out or test data set
Take the remaining groups as a training data set
Fit a model on the training set and evaluate it on the test set
Retain the evaluation score and discard the model
Summarize the skill of the model using the sample of model evaluation scores

```python
Y_predict
```

```python
# lets check accuracy for model
acc= accuracy_score(Y_test, Y_predict)
```

```python
acc # we got accuracy of 98 percent for our model
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, Y_predict)
plt.title('Logistic Regression Confusion Matrix')
sns.heatmap(cm,annot=True,fmt="d")
plt.show()# our model misclassified only 3 values out of 156 values and our model cant perform better than this stillw e will check with another classification algorithms
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(Y_test, Y_predict))
```

```python
# lets try random forest
```

even though our logistic model worked best on our dataset, we need to remeber it is a base model we tested with there are more versatile classification algoirthms we can actually put to good use in classifying data points correctly
in our logist model we misclassifed 2 legit firms as fraud
and 1 fraud firm as legit
in banking industry it is critically important for us to classify data poinst correctly we dont wanna lose credible banks just because our model classified them correctly nor we wanna have fraudelent firms in our list, its critically important and we use more advanced versatile machine learning algorithm to classify misclassified data points correctly

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators = 100,criterion='entropy', random_state = 50)
rf.fit(X_train_scaled, Y_train)
```

```python
acc_rf = cross_val_score(estimator = rf, X = X_train_scaled, y = Y_train, cv = kf, scoring=scoring)
acc_rf.mean() # we got an accuracy of 99 lets see if our model classifed all data points correctly
```

```python
Y_predict = rf.predict(X_test_scaled) 
# lets check accuracy for model
acc= accuracy_score(Y_test, Y_predict)
print(classification_report(Y_test, Y_predict))
```

```python
cm = confusion_matrix(Y_test, Y_predict)
plt.title('Random Forest Confusion Matrix')
sns.heatmap(cm,annot=True,fmt="d")
plt.show()
```

cm # bingo we classified all of our data points correctly , now we don't need to test any other machine learning classification model for our data as we classified 100 percent of data points correctly random forest wins

```python
# what if we want to check what are our top independent variables or features in our dataset aren't we curious to know what are those? lets check
```

```python
best_features = rf.feature_importances_
index = np.argsort(best_features)[::-1]
# lets rearrange our feature columns
names = [X.columns[i] for i in index]
plt.figure()
plt.title("Best Features")
plt.bar(range(X.shape[1]), best_features[index])
# Featurename
plt.xticks(range(X.shape[1]), names, rotation=90)
plt.show()
```
